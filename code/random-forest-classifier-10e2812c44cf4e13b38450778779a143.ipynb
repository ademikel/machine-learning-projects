```python
import pandas as pd
test_data = pd.read_csv("Test_Data.csv")
train_data = pd.read_csv("Train_Data.csv")
```

```python
from sklearn import preprocessing
encoder = preprocessing.LabelEncoder()

train_encoded_data = train_data.apply(encoder.fit_transform)
test_encoded_data = test_data.apply(encoder.fit_transform)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(train_encoded_data.iloc[:, :-1], train_encoded_data.iloc[:, -1])
predictions = model.predict(test_encoded_data.iloc[:, :])
```

```python
df = pd.DataFrame(predictions)
df.index = test_encoded_data.index
df.columns = ["prediction"]

# downloading locally
from google.colab import files
df.to_csv('prediction_results.csv')         
files.download('prediction_results.csv')
```
