```python
def fact(num):
    if num < 0:
        return 0
    elif num == 0 or num == 1:
        return 1
    else:
        init = 1
        while num > 1:
            init *= num
            num -= 1
        return init

n = int(input('Enter the value of n: '))
r = int(input('Enter the value of r: ')) 
print(fact(n) // fact(n-r))
```
